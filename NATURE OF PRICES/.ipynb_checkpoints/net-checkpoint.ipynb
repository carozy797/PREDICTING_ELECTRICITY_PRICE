```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

The Net Imbalance Volume- Price Relationship Display and Representaions

```python
#loading of data
df_net = pd.read_csv('net_imbalance.csv')

# make index the date and change it to Datetime instead of object
df_net = df_net.set_index('Settlement Date')
df_net.index = pd.to_datetime(df_net.index, dayfirst = True)

df_net.head()
```

```python
# dataset description
print("="*50)
print("Information about the dataset \n")
df_net.info()
print("\n")
print("="*50)
df_net.describe()
print("="*50)
print("Null values \n")
df_net.isnull().sum()
```

```python
# checking the relationship btwn net imbalance and the prices for settlement periods
plt.figure(figsize= (15,5))
x = df_net['Net Imbalance Volume(MWh)']
y_SSP = df_net['System Sell Price(GBP/MWh)']
plt.title("Net Imbalance vs SSP from 2019 - 18th May, 2023")

sns.scatterplot(x = x, y = y_SSP, color= 'red', label = 'SSP')
```

```python
plt.figure(figsize= (15,5))
x = df_net['Net Imbalance Volume(MWh)']
y_SBP = df_net['System Buy Price(GBP/MWh)']
plt.title("Net Imbalance vs SBP from 2019 - 18th May, 2023")

sns.scatterplot(x = x, y = y_SBP, color= 'blue', label = 'SBP')

plt.figure(figsize= (15,5))
sns.displot(x)

```

```python
plt.figure(figsize = (15, 5))
plt.ylabel("Prices")
plt.xlabel("Date")
plt.plot(df_net.index, y_SBP)
plt.title("Time series from 2019 - May 2023")
```

```python
plt.figure(figsize = (15, 5))
plt.title("Net imbalance volume series")
plt.ylabel("Net")
plt.xlabel("Date")
plt.scatter(df_net.index, x)
```

```python
# The nature of net imbalance for a day
plt.figure(figsize = (12,5))
day_2019 = x[: 48]
settlement_period = df_net['Settlement Period']
sns.lineplot(x = settlement_period[:48],y =  day_2019, color= 'red', label = 'day-1')
plt.title("Net Imbalance for a day")
plt.xlabel("48 hrs (Half an hour)")
plt.ylabel("Net Imbalance")

day2_2019 = x[48: 48+48]
sns.lineplot(x = settlement_period[48:48+48], y = day2_2019, color= 'blue', label = 'day-2')

day3_2019 = x[ 48+48: 48+48+48]
sns.lineplot(x = settlement_period[48+48:48+48+48], y = day3_2019, color= 'green', label = 'day-3')
```

```python
# The nature of net imbalance for a week
plt.figure(figsize = (12,5))
day_2019 = x[: 48*7]
settlement_period = df_net['Settlement Period']
sns.lineplot(x = settlement_period[:48*7],y =  day_2019, color= 'red')
plt.title("Net Imbalance for a week")
plt.xlabel("48 hrs (Half an hour)")
plt.ylabel("Net Imbalance")

 
```
