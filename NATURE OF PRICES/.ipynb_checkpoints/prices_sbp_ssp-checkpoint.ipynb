```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# reading the electricity price - system buying price and system support price
df_ssp = pd.read_csv('prices_ssp.csv')
df_ssp = df_ssp.drop(columns = ['49','50'])
df_sbp = pd.read_csv('prices_sbp.csv')
df_sbp = df_sbp.drop(columns = ['49','50'])
df_sbp.head()
```

```python
# There is the need to replace the nan
df_sbp = df_sbp.set_index('Date')
df_sbp.index = pd.to_datetime(df_sbp.index, dayfirst = True)


df_ssp = df_ssp.set_index('Date')
df_ssp.index = pd.to_datetime(df_ssp.index, dayfirst = True)



new_df_sbp = df_sbp.iloc[:, 1:]

df_sbp['47'].fillna(method = "bfill", inplace = True)
df_sbp['48'].replace(method = "bfill", inplace = True)
df_ssp['47'].replace(method = "bfill", inplace = True)
df_ssp['48'].replace(method = "bfill", inplace = True)

# removing the run column

new_df_sbp = df_sbp.iloc[:, 1:]
new_df_ssp = df_ssp.iloc[:, 1:]

new_df_ssp
```

```python
new_df_sbp_mean = new_df_sbp.mean(axis = 1)

new_df_sbp_mean
# new_df_ssp = df_ssp.iloc[:, 1:]
new_df_ssp_mean = new_df_ssp.mean(axis = 1)

new_df_ssp_mean
```

```python
plt.figure( figsize= (15, 6))
plt.title('plot of date and Avg prices (SSP, SBP) from 2001 - may, 2023')
plt.ylabel('Price')
sns.lineplot(x = new_df_ssp.index, y = new_df_ssp_mean,  color='blue', label='SSP')
sns.lineplot(x = new_df_sbp.index, y = new_df_sbp_mean,  color='red', label='SBP')
```
