Pricing data focusing on the SBP


Importing some important packages to be used

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Reading the data for processing and visualization

The date column is an object, so we make it a datetime object.

```python
# reading the electricity price - system buying price
df = pd.read_csv('prices_sbp.csv')
df = df.drop(columns = ['49','50'])

# make index the date and change it to Datetime instead of object
df = df.set_index('Date')
df.index = pd.to_datetime(df.index, dayfirst = True)
```

```python
# Displaying the data
df.head()
```

Huge data requires that we check its description instead of manually going through to check the errors and mistakes

```python
# dataset description
print("="*50)
print("Information about the dataset \n")
df.info()
print("\n")
print("="*50)
df.describe()
print("="*50)
print("Null values \n")
df.isnull().sum()
```

There is the need to replace the nan

settlement period 47 and 48 has 22 missing value

We decide to fill it with the previous price value, We assume that, 
there is an agreed price in each 30 min and that price 
not showing can affect our decision

```python
 
df['47'].fillna(method = "bfill", inplace = True)
df['48'].fillna(method = "bfill", inplace = True)
# taking away the run column
new_df = df.iloc[:, 1:]
new_df
```

```python
# checking if the nan values are now filled with the previous price
df.isna().sum()
```

```python
# data for each year from 2019 - 2023

data_2019 = new_df.iloc[6489: 6489+365 , :]
data_2020 = new_df.iloc[ 6489+365:  6489+365+366 , :]
data_2021 = new_df.iloc[ 6489+365+366: 6489+365+366+365, :]
data_2022 = new_df.iloc[ 6489+365+366+365:  6489+365+366+365+365, :]
data_2023 = new_df.iloc[6489+365+366+365+365: , :]


data_2022.head()
```

```python
# scatter  and line plot for functions
def plot_year(data, year):
    plt.figure(figsize=(12, 5))
    plt.title("Plot of " + str(year)+" prices for all data points")
    plt.xlabel('year')
    plt.ylabel('price')
    for i in range(1, 49):
        sns.lineplot(x = data.index, y = data[str(i)])
def sPlot_year(data, year):
    plt.figure(figsize=(12, 5))
    plt.title("Plot of " + str(year)+" prices for all data points")
    plt.xlabel('year')
    plt.ylabel('price')
    for i in range(1, 49):
        plt.scatter(data.index, data[str(i)])
def Plot_year_mean(data, year):
    plt.figure(figsize=(12, 5))
    plt.title("Plot of " + str(year)+" prices for mean or Avg data points")
    plt.xlabel('year')
    plt.ylabel('price')
    for i in range(1, 49):
        plt.scatter(data.index, data.mean(axis=1))

def plot_show_category(data, year):
    # Set the Seaborn style
    sns.set(style="darkgrid")
    plt.figure(figsize=(12, 5))
    

    dfs = []

    for i in range(1, 49):


        # Create a DataFrame for the current category with 'x' and 'y' values
        df_category = pd.DataFrame({'x': data.index, 'y':  data[str(i)], 'half-hour': i})

        # Append the current category DataFrame to the list
        dfs.append(df_category)

    # Concatenate the data frames in the list into a single DataFrame
    df = pd.concat(dfs)

    # Get the number of unique categories
    num_categories = df['half-hour'].nunique()

    # Set a color palette with a different color for each category
    colors = sns.color_palette('Set1', n_colors=num_categories)

    # Plot the time series graph with hue representing the 48 categories
    sns.lineplot(data=df, x='x', y='y', hue='half-hour', palette=colors)
    plt.title("Plot of " + str(year) + "prices for all data points")
    plt.xlabel('year')
    plt.ylabel('price')
    # Display the plot
    plt.show()
        
```

```python
plt.figure( figsize= (15, 6))
plt.title("Date vs price 2001 - 2023 : For all data points (Not Avg for a day)")
plt.xlabel("Date")
plt.ylabel("Price")
for i in range(1, 49):
    plt.plot(new_df.index, new_df[str(i)])
```

```python
# Assuming 'data_2023_05' is a pandas DataFrame with columns representing each half-hourly data point
plt.figure( figsize= (15, 8))
# Set the Seaborn style
sns.set(style="darkgrid")

# Set the x-axis values for the 48 half-hourly data points
x = np.arange(1, 49)

# Create an empty list to store the data frames for all seven categories
dfs = []
days = ['day-1', 'day-2', 'day-3', 'day-4', 'day-5', 'day-6', 'day-7']

# Loop over the seven categories
for i in range(7):
    # Choose the desired row 'i' from 'data_2023_05'
    data_transpose = data_2001_05.iloc[i, :]
    
    # Create a DataFrame for the current category with 'x' and 'y' values
    df_category = pd.DataFrame({'x': x, 'y': data_transpose, 'days': days[i]})
    
    # Append the current category DataFrame to the list
    dfs.append(df_category)

# Concatenate the data frames in the list into a single DataFrame
df = pd.concat(dfs)

# Get the number of unique categories
num_categories = df['days'].nunique()

# Set a color palette with a different color for each category
colors = sns.color_palette('Set1', n_colors=num_categories)

# Plot the time series graph with hue representing the seven categories
sns.lineplot(data=df, x='x', y='y', hue='days', palette=colors)

# Set the labels and title for the graph
plt.xlabel('2001-05 (half an hr)')
plt.ylabel('prices')
plt.title('First week in May 2001: Time series plot for 48 hrs data points (half an hour)')

# Display the plot
plt.show()
```
